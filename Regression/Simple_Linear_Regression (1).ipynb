```python
import numpy as np

# Step 1: Get the Dataset
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Step 2: Compute the Mean of X and y
mean_X = np.mean(X)
mean_y = np.mean(y)

# Step 3: Calculate Coefficients
m = np.sum((X - mean_X) * (y - mean_y)) / np.sum((X - mean_X)**2)
c = mean_y - m * mean_X

# We have our model
print(f"Model: y = {m}*x + {c}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 6))
plt.scatter(X, y, color = "red", marker = "o", s = 30)
y_pred = m * X + c
plt.plot(X, y_pred, color = "blue")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Linear Regression")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define a function to calculate mean and coefficients
def compute_coefficients(X, y):
    mean_X = np.mean(X)
    mean_y = np.mean(y)

    # Calculate slope (m)
    numerator = np.sum((X - mean_X) * (y - mean_y))
    denominator = np.sum((X - mean_X) ** 2)
    m = numerator / denominator

    # Calculate intercept (c)
    c = mean_y - m * mean_X
    return m, c

# Step 2: Define a function to make predictions
def predict(X, m, c):
    return m * X + c

# Step 3: Define a function to evaluate model using R² score
def evaluate_model(y_true, y_pred):
    ss_total = np.sum((y_true - np.mean(y_true)) ** 2)
    ss_residual = np.sum((y_true - y_pred) ** 2)
    r2 = 1 - (ss_residual / ss_total)
    return r2

# Step 4: Define a function to plot the results
def plot_regression(X, y, y_pred):
    plt.figure(figsize=(6, 6))
    plt.scatter(X, y, color="red", label="Actual Data")
    plt.plot(X, y_pred, color="blue", label="Regression Line")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.title("Simple Linear Regression")
    plt.legend()
    plt.grid(True)
    plt.show()

# Step 5: Main logic
def run_linear_regression(X, y):
    m, c = compute_coefficients(X, y)
    y_pred = predict(X, m, c)
    r2 = evaluate_model(y, y_pred)

    print(f"Model: y = {m:.2f}*x + {c:.2f}")
    print(f"R² Score: {r2:.4f}")
    plot_regression(X, y, y_pred)

# Example Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Run the regression
run_linear_regression(X, y)
```
